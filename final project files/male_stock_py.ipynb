```python
tickers = ['BURL','CFG','Y',"INTC","MSFT"]
query_url = "https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol="
api_key = "G3HAJBMGFTXEOWHT"
male_stock_response = []

for x in range(len(tickers)):

    print(f"Making request number: {x} for : {tickers[x]}")

    # Get one of the posts
    male_post_response = requests.get(f"{query_url}{tickers[x]}&apikey={api_key}").json()
    
    # Save post's JSON
    male_stock_response.append(male_post_response)
```

```python
male_tickers2 = ['UTX','RTN','QRTEA','F','TJX']
for x in range(len(male_tickers2)):

    print(f"Making request number: {x} for : {male_tickers2[x]}")

    # Get one of the posts
    male_post_response = requests.get(f"{query_url}{male_tickers2[x]}&apikey={api_key}").json()
    
    # Save post's JSON
    male_stock_response.append(male_post_response)
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(male_stock_response)
data = male_stock_response
```

```python
def flatten_json(z):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], str(name) + str(a) + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, str(name) + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(z)
    return out


flattened_dict_male = [flatten_json(d) for d in data]
flattened_dict_male
```

```python
male_stock_df = pd.DataFrame(flattened_dict_male)
male_stock_df
```

```python
male_stock_df = male_stock_df.drop(columns=['Meta Data_1. Information', 'Meta Data_3. Last Refreshed','Meta Data_4. Time Zone'])
male_stock_df
```

```python
melted_male_stock = male_stock_df.melt(id_vars=["Meta Data_2. Symbol",], 
        var_name="Date", 
        value_name="Value")

melted_male_stock.head()
```

```python
# new data frame with split value columns 
new2 = melted_male_stock["Date"].str.split("_", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_male_stock["Old"]= new2[0] 
  
# making separate last name column from new data frame 
melted_male_stock["New Date"]= new2[1] 

  
# Dropping old Name columns 
melted_male_stock.drop(columns =['Old'], inplace = True)

melted_male_stock.head()
```

```python
# new data frame with split value columns 
new2 = melted_male_stock["New Date"].str.split(".", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_male_stock["New Date"]= new2[0] 
  
# making separate last name column from new data frame 
melted_male_stock["Stock Data Type"]= new2[1] 

  
# Dropping old Name columns 

melted_male_stock.head()
```

```python
# new data frame with split value columns 
new2 = melted_male_stock["New Date"].str.split("_", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_male_stock["New Date"]= new2[0] 
  
# making separate last name column from new data frame 
melted_male_stock["Drop"]= new2[1] 
```
