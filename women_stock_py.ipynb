Run API to gather stock data 5 tickers at a time (only 5 call per minute allowed through alphavantage)

```python
tickers = ['WSM', 'KEY','PGR','IBM','ORCL']
query_url = "https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol="
api_key = "G3HAJBMGFTXEOWHT"
stock_response = []

for x in range(len(tickers)):

    print(f"Making request number: {x} for : {tickers[x]}")

    # Get one of the posts
    post_response = requests.get(f"{query_url}{tickers[x]}&apikey={api_key}").json()
    
    # Save post's JSON
    stock_response.append(post_response)
```

```python
tickers2 = ['LMT','GD','ULTA','GM','ROST']
for x in range(len(tickers2)):

    print(f"Making request number: {x} for : {tickers2[x]}")

    # Get one of the posts
    post_response = requests.get(f"{query_url}{tickers2[x]}&apikey={api_key}").json()
    
    # Save post's JSON
    stock_response.append(post_response)
```

Print stock data to pretty json

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(stock_response)
data = stock_response
```

use flatten_json to flatten the heavily nested dictionaries 

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], str(name) + str(a) + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, str(name) + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out


flattened_dict = [flatten_json(d) for d in data]
flattened_dict
```

```python
stock_df = pd.DataFrame(flattened_dict)
stock_df
```

```python
stock_df = stock_df.drop(columns=['Meta Data_1. Information', 'Meta Data_3. Last Refreshed','Meta Data_4. Time Zone'])
stock_df
```

continue cleaning and melting data to create desired data table

```python
melted_stock = stock_df.melt(id_vars=["Meta Data_2. Symbol",], 
        var_name="Date", 
        value_name="Value")

melted_stock.head()
```

split cells to isolate desired index data

```python
# new data frame with split value columns 
new2 = melted_stock["Date"].str.split("_", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_stock["Old"]= new2[0] 
  
# making separate last name column from new data frame 
melted_stock["New Date"]= new2[1] 

  
# Dropping old Name columns 
melted_stock.drop(columns =['Old'], inplace = True)

melted_stock.head()
```

```python
# new data frame with split value columns 
new2 = melted_stock["New Date"].str.split(".", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_stock["New Date"]= new2[0] 
  
# making separate last name column from new data frame 
melted_stock["Stock Data Type"]= new2[1] 

  
# Dropping old Name columns 

melted_stock.head()
```

```python
# new data frame with split value columns 
new2 = melted_stock["New Date"].str.split("_", n = 1, expand = True) 
  
# making separate first name column from new data frame 
melted_stock["New Date"]= new2[0] 
  
# making separate last name column from new data frame 
melted_stock["Drop"]= new2[1] 
```

```python
new_stock = melted_stock.drop(columns = ["Drop","Date"])
```

```python
new_stock['index1'] = new_stock.index
```

```python
new_stock2 = new_stock.copy()
new_stock2['idx'] = new_stock2['Meta Data_2. Symbol'].astype(str) + '__' + new_stock2['New Date'].astype(str)
new_stock2 = new_stock2.set_index('idx')
new_stock2.head()
```

```python
pivoted = new_stock2.pivot(columns='Stock Data Type', values='Value')
pivoted = pivoted.reset_index(drop=False)
pivoted.head()
```

```python
# new data frame with split value columns 
newp = pivoted["idx"].str.rsplit("_", n = 1, expand = True) 
  
# making separate first name column from new data frame 
pivoted["Drop"]= newp[1] 
  
# making separate last name column from new data frame 
pivoted["Ticker"]= newp[0] 
```
