```python
import json
import requests
import csv
import pandas as pd
import pprint
from pandas.io.json import json_normalize
```

```python
women_CEOs_path = "2019_WomenCEO_list.csv"
women_CEOs = pd.read_csv(women_CEOs_path)
women_CEOs
```

```python
tickers = women_CEOs["Ticker Symbol (NYS/NASDAQ)"]
query_url = "https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol="
api_key = "5KPEJAL4QA90XTN1"
stock_response = []

for x in range(len(tickers)):

    print(f"Making request number: {x} for : {tickers[x]}")

    # Get one of the posts
    post_response = requests.get(f"{query_url}{tickers[x]}&apikey={api_key}").json()
    
    # Save post's JSON
    stock_response.append(post_response)
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(stock_response)
data = stock_response
len(stock_response)
```

```python

result = json_normalize(data, 'Meta Data')

result
```

```python
count = 0

for key in stock_response[count]['Monthly Time Series']:
    result = json_normalize(data, key)
         
count += 1

result
```
